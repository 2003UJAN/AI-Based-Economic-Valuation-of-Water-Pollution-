```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import lime
import lime.lime_tabular
import shap
```

```python
# Load the data
df = pd.read_csv('/content/india_water_pollution_economic_timeseries.csv')
```

```python
# Convert 'Year' to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Set 'Year' as the index
df.set_index('Year', inplace=True)
```

```python
# Display basic statistics
print(df.describe())
```

```python
# Plot histograms for key variables
key_vars = ['pH', 'TDS (mg/L)', 'Heavy Metals (mg/L)', 'BOD (mg/L)', 'Cholera Cases (%)', 'Typhoid Cases (%)', 'Dysentery Cases (%)']
df[key_vars].hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Plot time series for pH levels
plt.figure(figsize=(12, 6))
for city in df['Region'].unique():
    city_data = df[df['Region'] == city]
    plt.plot(city_data.index, city_data['pH'], label=city)
plt.title('pH Levels Over Time by City')
plt.legend()
plt.show()
```

```python
# Plot correlation heatmap
correlation_matrix = df[key_vars + ['Healthcare Cost (INR per Capita)', 'GDP Loss Due to Water Pollution (INR Millions)']].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Prepare data for LSTM
city_data = df[df['Region'] == 'Mumbai']['pH'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(city_data)
```

```python
# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:(i + seq_length), 0])
        y.append(data[i + seq_length, 0])
    return np.array(X), np.array(y)

seq_length = 5
X, y = create_sequences(scaled_data, seq_length)
```

```python
# Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
# Reshape input for LSTM [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# Build LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(seq_length, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=32, verbose=0)
```

```python
# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
# Invert predictions to original scale
train_predict = scaler.inverse_transform(train_predict)
y_train = scaler.inverse_transform([y_train])
test_predict = scaler.inverse_transform(test_predict)
y_test = scaler.inverse_transform([y_test])
```

```python
# Calculate RMSE
train_rmse = np.sqrt(np.mean((train_predict - y_train)**2))
test_rmse = np.sqrt(np.mean((test_predict - y_test)**2))
print(f'Train RMSE: {train_rmse}, Test RMSE: {test_rmse}')
```

```python
# Plot LSTM results
plt.figure(figsize=(12, 6))
plt.plot(y_test[0], label='Actual')
plt.plot(test_predict, label='Predicted')
plt.title('LSTM: Actual vs Predicted pH Levels for Mumbai')
plt.legend()
plt.show()
```

```python
#Prepare data for ARIMA
city_data = df[df['Region'] == 'Mumbai']['pH']
```

```python
# Fit ARIMA model
model = ARIMA(city_data, order=(1, 1, 1))
results = model.fit()
```

```python
# Make predictions
predictions = results.forecast(steps=len(city_data))
```

```python
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(city_data, predictions))
print(f'ARIMA RMSE: {rmse}')
```

```python
# Plot ARIMA results
plt.figure(figsize=(12, 6))
plt.plot(city_data.index, city_data, label='Actual')
plt.plot(city_data.index, predictions, label='Predicted')
plt.title('ARIMA: Actual vs Predicted pH Levels for Mumbai')
plt.legend()
plt.show()
```

```python
# Prepare data for XAI
X = df[key_vars].values
feature_names = key_vars
```

```python
# LIME
explainer = lime.lime_tabular.LimeTabularExplainer(X, feature_names=feature_names, class_names=['pH'], verbose=True, mode='regression')
```

```python
instance_df = pd.DataFrame([X[0]], columns=feature_names)

def predict_fn(data):

    data = pd.DataFrame(data, columns=feature_names)

    predictions = results.forecast(steps=len(data))

    return predictions.values

exp = explainer.explain_instance(instance_df.iloc[0], predict_fn, num_features=len(feature_names))
exp.show_in_notebook(show_table=True)
```

```python
# SHAP
def predict_fn(data):
    data = data.reshape(-1, X.shape[1])  # Ensure proper reshaping
    data = pd.DataFrame(data, columns=key_vars)  # Convert to DataFrame

    # Make predictions using the model
    predictions = results.predict(start=len(city_data),
                                  end=len(city_data) + len(data) - 1,
                                  dynamic=False,
                                  exog=data)
    return predictions

# Reduce background samples to speed up SHAP computations
background = shap.sample(X, 100)  # Select a representative subset of 100 samples

# Create SHAP KernelExplainer with reduced background
explainer = shap.KernelExplainer(predict_fn, background)

# Compute SHAP values
shap_values = explainer.shap_values(X)

# Generate SHAP summary plot
shap.summary_plot(shap_values, X, feature_names=feature_names)
```
